## Last homework
Solve the same equations using `scipy.integrate.odeint` and `scipy.integrate.solve_ivp`

$$
\begin{align*} 
\frac{dx}{dt}&=\sigma (y-x) \\
\frac{dy}{dt}&=x(\rho-z) -y \\
\frac{dz}{dt}&=xy-\beta z 
\end{align*} 
$$

for the parameters:

* $\sigma =10$
* $\rho=28$
* $\beta=8/3$

lets set intial conditions:

* x(0)=1.0
* y(0)=1.0
* z(0)=1.0

Compare the results. Seeing anything strange?

```python
import numpy as np
from scipy.integrate import odeint
from scipy.integrate import solve_ivp
import matplotlib.pyplot as plt
```

## scipy.integrate.odeint

```python
def odes(ic,t,sigma,rho,beta):
  x, y, z = ic[0], ic[1], ic[2]  # Condiciones iniciales para cada variable

  ode1 = sigma*(y-x) # dx/dt
  ode2 = x*(rho-z) - y  # dy/dt
  ode3 = x*y - beta*z  # dz/dt

  return [ode1,ode2,ode3] 
```

```python
#Parámetros
sigma = 10
rho = 28
beta = 8/3

# Tiempos de cálculo
t = np.linspace(0.0,20.0,101)

# Condiciones iniciales
ic = [1.0,1.0,1.0]
```

```python
solucion = odeint(odes,ic,t,args=(sigma,rho,beta))

xsol = solucion[:,0]
ysol = solucion[:,1]
zsol = solucion[:,2]
```

```python
fig,ax=plt.subplots(1,3,figsize=(14,3))

ax[0].plot(t,xsol,'-r')
ax[0].set_title('Evolución de x en el tiempo')
ax[0].set_xlabel('Tiempo')
ax[0].set_ylabel('x')

ax[1].plot(t,ysol,'-b')
ax[1].set_title('Evolución de y en el tiempo')
ax[1].set_xlabel('Tiempo')
ax[1].set_ylabel('y')

ax[2].plot(t,zsol,'-g')
ax[2].set_title('Evolución de z en el tiempo')
ax[2].set_xlabel('Tiempo')
ax[2].set_ylabel('z')
```

##  scipy.integrate.solve_ivp

```python
def derivs(t,ci,sigma,rho,beta):
  x, y, z = ci[0], ci[1], ci[2]  # Condiciones iniciales para cada variable

  edo1 = sigma*(y-x) # dx/dt
  edo2 = x*(rho-z) - y  # dy/dt
  edo3 = x*y - beta*z  # dz/dt

  return [edo1,edo2,edo3]
```

```python
#Parámetros
sigma = 10
rho = 28
beta = 8/3

# Intervalo de tiempo
t_span = (0.0,20.0)

# Condiciones iniciales
ci = [1.0,1.0,1.0]
```

```python
sol = solve_ivp(derivs,t_span,ci,args=(sigma,rho,beta))
```

```python
fig,ax=plt.subplots(1,3,figsize=(14,3))

ax[0].plot(sol.t,sol.y[0],'-r')
ax[0].set_title('Evolución de x en el tiempo')
ax[0].set_xlabel('Tiempo')
ax[0].set_ylabel('x')

ax[1].plot(sol.t,sol.y[1],'-b')
ax[1].set_title('Evolución de y en el tiempo')
ax[1].set_xlabel('Tiempo')
ax[1].set_ylabel('y')

ax[2].plot(sol.t,sol.y[2],'-g')
ax[2].set_title('Evolución de z en el tiempo')
ax[2].set_xlabel('Tiempo')
ax[2].set_ylabel('z')
```
