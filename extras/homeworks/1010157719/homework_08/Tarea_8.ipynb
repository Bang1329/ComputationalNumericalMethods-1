a) Implement the above procedure numerically to find y(x) from:
$$ \frac{dy}{dx} = yx^2 - 1.2y$$


 
- The initial condition is: y(0)=1 over the interval [0,2]
- The range of x is actually the interval [0,2]
- Plot the y(x) for h=0.5
- On the same plot compare it with the analytical solution of this differential equation, which is:
$$y(x) = e^{x^3/3-1.2x}$$


```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def func(x,y): #dy/dx
  return y*x**2 - 1.2*y
```

```python
def fa(x): #Solución Analítica 
  return np.exp(x**3 /3 - 1.2*x)
```

```python
def euler_int(func, intrange, h, yinit): #implementación numérica para solucionar la ecuación
  x = np.arange(intrange[0], intrange[1] + h, h)
  y = np.zeros(len(x)+1)
  y[0] = yinit
  for j,i in enumerate(x):
    y[j+1] += [y[j] + h*func(i,y[j])]

  x = np.append(x,x[-1] + h) 
  return x, y
```

```python
intrange=[0,2.0] #rango  
h=0.5 #paso 
yinit=1 #Condición Inicial
x1,y1=euler_int(func,intrange,h,yinit)

fig,ax=plt.subplots(figsize=(8, 6), dpi=80)
ax.set_title('Gráfica Ecuación Diferencial')
ax.plot(x1,y1,'-r',label='Solución Numérica')
ax.plot(x1[0:5], fa(x1[0:5]), label = 'Solución Analítica')
ax.grid()
ax.legend()
```
