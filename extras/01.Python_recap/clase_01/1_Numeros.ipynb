## Python como una calculadora

Python se puede utilizar como una calculadora simple, se puede introducir una expresión y este escribirá los valores. La sintaxis es sencilla: los operadores `+ , -, "*", "**", /, //, %`  y los paréntesis `(`, `)` pueden ser usados para agrupar los operaciones. [1]

````python
# operación de suma
9876 + 2345 # suma de dos números enteros
````

````python
(1000 - 3*7) / 8 # operaciones mixtas
````

Tenga en cuenta que el resultado de la primera operación devolvió un número entero, mientras que la segunda operación devolvió un `float`. Los números enteros no tienen parte fracional (ej. 2, 4, 20) y se llaman en python `int`; los que tienen una parte fraccionaria (ej. 5.0, 1.6) son el tipo `float`. Más adelante se abordarán de los tipos de números en el tutorial.
Por su parte, la división `/` siempre retorna un punto flotante. Para hacer floor division y obtener un resultado entero (descartando cualquier resultado fraccionario) puede usarse el operador `//`.

````python
47 / 7 # la división clásica devuelve un flotador
````

````python
47 // 7 # la división de piso descarta la parte fraccionaria
````

Para calcular el resto de la división se puede usar el operador módulo: `%`.  El operador módulo es muy útil para determinar si un número es par o impar. Usando el operador módulo con un módulo de 2, se puede verificar cualquier número para ver si es uniformemente divisible por 2. Si es uniformemente divisible (el resto de la división es cero), entonces es un número par, y si no es uniformemente divisible (el resto de división es uno), es un número impar. A continuación dos ejemplos: 

````python
24 % 2 # El resultado de usar operador módulo: 
       # 24%2 es 0 porque 24 es número par)
````

````python
25 % 2 # El resultado de usar operador módulo: 
       # 25%2 es 1 porque 25 es número impar)
````

````python
25**25 # 25 a la potencia 25
````

##  Operadores de comparación y operadores lógicos

Los operadores de comparación se utilizan para comparar dos valores y dar un resultado de `True` o `False` (booleanos).

    Símbolo:  Significado:    

    ==        igual          
    !=        no igual       
    >         mayor          
    <         menor         
    >=        mayor o igual 
    <=        menor o igual 

Los operadores de comparación se pueden usar en conjunto con operadores lógicos.

    and       ambas expresiones son verdad: True
    or        por lo menos una expresión es verdad: True  
    not       expresión no es verdad: not True

````python
int_1=10            # se define una variable int_1 que es número entero con el valor 10
int_2=20            # se define otra variable int_2 que es número entero con el valor 20
print(int_1==int_2) # se "pregunta" si son iguales, que es falso (False)
print(int_1!=int_2) # se "pregunta" si son diferentes, que es verdad (True)
````

````python
print(int_1>1) # se "pregunta" si la primer variable es mayor que 1, que es verdad (True)
````

````python
print(int_1%2==0) # se "pregunta" si la variable int_1 es par, que es verdad (True)
````

````python
((int_1!=int_2) and (int_1>1)) # se "pregunta" si la primer variable es diferente que el segundo (True)
                               # Y si la primer variable es mayor que 1 (True)
                               # ambas respuestas son True , entonces resultado final es True
````

````python
((int_1!=int_2) and (int_1>100)) # se "pregunta" si la primer variable es diferente que el segundo (True)
                                 # Y si la primer variable es mayor que 100 (False)
                                 # solo una respuesta es True, entonces resultado final es Falso
                                 # (operador and requiere que ambos resultados son verdad para devolver True )
````

````python
((int_1!=int_2) or (int_1>100))  # se "pregunta" si la primer variable es diferente que el segundo (True)
                                 # O si la primer variable es mayor que 100 (False)
                                 # el resultado es True porque el operador or requiere que por lo menos un resultado sea verdad para devolver True
````

````python
((int_1!=int_2) or not (int_1>100))  # se "pregunta" si la primer variable es diferente al segundo (True)
                                 # and si la primer variable NO es mayor que 100 (True)
                                 # el resultado es True porque el operador de las variables es diferente, ya que int_1 NO es mayor que 100
````

````python
# cada uno de estos tipos de operaciones lógicas que devuelve valor tipo bool puede ser guardada 
# en otro variable (y/o imprimida)
verdad_o_falos = ((int_1!=int_2) or not (int_1>100))
print(verdad_o_falos)
````

## Expresiones condicionales, if, else, elseif

La estructura de la expresión condicional tipo: `if - else`, es la siguiente:
```
if (operación de comparación):  
    ejecutar aquí si resultado es True  
else:  
    ejecutar aquí si resultado es False  
```        
        
Existen dos elementos importantes 
* los dos puntos `:` después de la operación de comparación
* la indentación de expresiones que se van a ejecutar en cualquier caso  
(En Python la indentación se hace usando `4` espacios (menos común `2` espacios) y es obligatoria en caso de usar todas las heramientas de flujo tipo: expresiones condicionales o loops (ciclos))

Abajo se presentan algunos ejemplos al usar `if, else, not, or, and`.
Es recomendable estudiar y entender los ejemplos antes de seguir avanzando con el material.

````python
# en el ejemplo siguiente se definen dos variables 
# y usando una expresión condicional se define una accion dependiente 
# del resultado de comparación

a = 200
b = 100

if (b > a):                         # si comparacion b > a devuelve valor TRUE
    print("b es major que a")       # imprime información: "b es mayor que a"
else:                               # si comparación b>0 devuelve False
    print("a es major o igual b")   # imprime información: "a es mayor que b"
    
# en este caso particular b no es mayor que a, entonces el resultado de la informacion es:
# a es mayor o igual b  
````

````python
# en el ejemplo siguiente se definen dos variables 
# y usando una expresión condicional se define una acción dependiente 
# del resultado de comparación
# se percibe que b es aumentado

a = 2000
b = 1000000

if (b > a):                         # si comparación b > a devuelve valor TRUE
    print("b es major que a")       # imprime información: "b es major que a"
else:                               # si comparación b>a devuelve False
    print("a es major o igual b")   # imprime información: "a es mayor o igual que b"
                                    # porque si b es igual a entonces b>a es falso

# en este caso particular b es mayor que a, entonces el resultado de la información es:
# "b es mayor que a"   
````

````python
# en el ejemplo siguiente se verifican dos condiciones conjunto con operador logico AND:
# si b es mayor que a (AND) si b es mayor que a multiplicado por 100

a = 2000
b = 1000000

if (b > a) and (b > a*100):           # si ambas comparaciones (b > a) AND (b > a) devuelven valor TRUE
    print("b es mucho major que a")   # imprime información: "b es mucho mayor que a"
else:                                 # en otro caso
    print("a es major que b")         # imprime información: "a es mayor que b"
    
# en este caso particular b es 1000000 que es mucho major que a
````

````python
# en el ejemplo siguiente se usa inicialmente función isinstance() 
# para determinar si ambos variables a y b son numéricos.
# el valor non-numerico sería, por ejemplo, una cadena de letras como 'abc' 
# "not isinstance(a,str)) vuelve True si "a" NO es tipo cadena de letras.
# "not isinstance(b,str)) vuelve True si "b" NO es tipo cadena de letras.
# si ambos son True operación de comparacion los numéricos puede comenzar

a = 200
b = "abc"

if (not isinstance(a,str)) and (not isinstance(b,str)):
    if (b > a):
        print("b es major que a")
    else:
        print("a es major o igual b")
else:
    print('a o b no tiene valor numerico')
````

Si se necesitan más de dos opciones (`if / else`) se usan bloques adicionales `elif`. La opción `elif` permite escoger una única opción entre varios. La palabra reservada `elif` es una abreviación de `else if`, y es útil para evitar un sangrado excesivo. [4]

````python
# En el siguiente ejemplo para dado valor de puntaje numérico, se califica al valor
# a una categoria A,B,C o D

score=70
if score>90:
    print('A')
elif score>80 and score<=90:
    print('B')    
elif score>=60 and score<=80:
    print('C')    
else:
    print('D')
````

## Loops: `for`, `while` y función `range()`
Algunas veces en programación es necesario escribir un conjunto de instrucciones repetidamente, lo cual es dispendioso. Por lo tanto, se utiliza la técnica de iteración para repetir el mismo tipo o tipos similares de tareas en función de la condición especificada. En tal sentido, una instrucción cíclica (en inglés: `loop`) permite ejecutar una instrucción o grupo de instrucciones varias veces. 

#### for 

En el loop `for` el número de repeticiones para el valor numérico se puede definir usando la función range(). La función `range(start,stop,step) devuelve una secuencia de números a partir de 0 (por defecto) e incrementa en 1 (por defecto), y termina en un número especificado excluido.

Por ejemplo: 
* `range(5)` especifica el rango de números 0,1,2,3,4 (por defecto el primer número es 0 y el último número es 5 excluido)
Para cambiar el primer número de rango (por defecto 0): 
* `range(1,6)` especifica el rango de números 1,2,3,4,5 (primer número 1 hasta 6, último excluido)
* `range(1,1)` no especifica ningún número
* `range(3,10)` especifica el rango de números 3,4,5,6,7,8,9 (primer número 3, último número 10 es excluido)
Para cambiar incremento que por defecto es 1: 
* `range(0,11,2)` el rango de números 0,2,4,6,8,10 (incremento es 2)
* `range(3,13,3)` el rango de números 3,6,9,12 (incremento es 3)
* `range(5,0,-1)` el rango de números 5,4,3,2,1 (incremento es -1, el último número no está incluido)

* `range()` acepta solo números enteros

En Python, las secuencias cíclicas tienen la siguiente estructura:
```
for x in range(argumento):  
    declaracion   
```
Nótese que:
* `for` es una palabra reservada y no puede cambiarse.
* La palabra (letra) `x` es un contador (variable) que puede tener cualquier nombre.
* `in range` significa adentro del rango especificado por argumento.
* `:` dos puntos al fin de la primera línea.
* A continuación la declaración se ejecutará varias veces. Tenga en cuenta que cuatro espacios antes de la declaración son obligatorios.

Entender esta estructura es más fácil con un ejemplo:

````python
for x in range(5):  # este loop va a ser ejecutado 5 veces
     print(x)       # cada vez x va a tener diferente valor: primero 0, después 1, después 2, después 3, después 4 
                    # que se puede imprimir x
````

````python
for x in range(3):        # este loop va a ser ejecutado 3 veces
     print('hola')        # cada vez x va a tener diferente valor (0,1,2) pero no necesitamos usarlo
                          # se puede solamente imprimir tres veces una cadena de letras (text): 'hola'
````

````python
for x in range(3):        # este loop va a ser ejecutado 3 veces
     print('hola', x)     # cada vez x va a tener diferente valor (0,1,2) que se puede imprimir juntos con cadena de letras
````

````python
for x in range(0,6): # aquí x va cambiar valores a través (0,1,2,3,4,5)
    print(x**2)      # imprimir x**2
````

Loop `for` puede ser útil para realizar una suma. Un ejemplo sencillo sería:

$$1+2+3+4+5\cdots+10=\sum_{i=1}^{10}i$$ 

````python
sum=0                 # se declara una variable sum e iniciaremos con valor 0
for i in range(1,11): # aquí x va cambiar valores a través (0,1,2,3..9) - 10 no está incluido
    sum=sum+i         # suma está aumentandose con cada valor nuevo de `x` 
    print(sum)        # imprimir resultado
````

Se puede imprimir el resultado y el mensaje al mismo tiempo:

````python
for x in range(0,6):                  # aquí x va cambiar valores a través (0,1,2,3,4,5)
    print(x, 'elevado a 2 es', x**2)  # imprimir valor de x, texto: "elevado a 2 es", valor x**2
````

````python
# siguiente loop no va a ser ejecutado ni una vez porque el rango range(1,1) está vacío
for x in range(1,1):
    print(x)
````

Los loops pueden ser jerarquizados (en inglés: nested) que significa loop adentro de loop. Es importante aprender a leer cómo se ejecuta secuencialmente el loop. Por ejemplo:

````python
# en el siguiente ejemplo hay dos loops, 
# primera loop, y dentro segunda loop

for x in range(0,4):
    print('x=',x)  
    for y in range(0,3):
        print('y=',y)

# primera loop for x=0: -> print: x=0, segunda loop for: y=0 -> print: y=0, y=1 -> print: y=1, y=2 -> print: y=2
# primera loop for x=1: -> print: x=1, segunda loop for: y=0 -> print: y=0, y=1 -> print: y=1, y=2 -> print: y=2
# primera loop for x=2: -> print: x=2, segunda loop for: y=0 -> print: y=0, y=1 -> print: y=1, y=2 -> print: y=2
# primera loop for x=3: -> print: x=3, segunda loop for: y=0 -> print: y=0, y=1 -> print: y=1, y=2 -> print: y=2
````

A continuación, otro ejemplo más interesante en el que se utiliza el contador de un loop en el segundo loop:

````python
# en el siguiente ejemplo hay dos loops, 
# primera loop, y dentro segunda loop

for x in range(1,7):  
    for y in range(1,x):
        print(y)

# primera loop x=1: -> segunda loop: no ejecutada!
# primera loop x=2: -> segunda loop: y=1 -> print: 1   
# primera loop x=3: -> segunda loop: y=1 -> print: 1, y=2 -> print: 2   
# primera loop x=4: -> segunda loop: y=1 -> print: 1, y=2 -> print: 2, y=3 -> print: 3 
# primera loop x=5: -> segunda loop: y=1 -> print: 1, y=2 -> print: 2, y=3 -> print: 3, y=4 -> print: 4...  
````

#### while
 En Python existe otra forma de ejecutar un bloque repetidamente. 
`while` permite ejecutar un bloque de instrucciones repetidamente hasta que se cumple una condición dada. Y cuando la condición se vuelve falsa, se ejecuta la línea inmediatamente después del bucle en el programa. 

La estructura del loop while es la siguiente:
```
while (expresión que vuelve valor booleano):
    declaración
```
El loop `while` necesita un incrementador para terminar la ejecución.
Un ejemplo siguiente muestra el funcionamiento de loop `while`:

````python
# while puede funcionar de la misma manera que "for" (comparar con ejemplo anterior usando loop "for")
i = 0            # un incrementador i se inicia con valor 0
while (i < 3):   # while verifica si i<3, es verdad (True) porque 0<3, entonces se ejecuta siguiente liean indentada
    print(i)     # imprime valor de contador: 0
    i = i + 1    # valor de incrementador se incrementa hasta 1
                 # loop vuelve a while que verifica si i<3, que es verdad (True) porque 1<3
                 # imprime valor de contador: 1
                 # valor de incrementador se incrementa hasta 2
                 # loop vuelve a while que verifica si i<3, que es verdad (True) porque 2<3
                 # imprime valor de contador: 2
                 # valor de incrementador se incrementa hasta 3  
                 # loop vuelve a while que verifica si i<3, que no es verdad (False) porque 3==3
                 # el bloque indentado no se ejecuta más 
                 # se ejecuta la siguiente parte de codigo afuera de bloque    
print('el valor final de incrementador es: ',i)  # imprime valor final de incrementador
````

*hay otra forma alternativa de hacer incremento:
`i+=1`

A veces no se sabe cuántas iteraciones debe hacer el bucle. En ese caso es útil crear un bucle "infinito" que pueda salir cuando se cumpla alguna condición. 

El loop `while` puede ser infinito si la expresión es siempre verdad. 
Para terminar el loop infinito se usa la palabra `break` dada otra condición (`if,else`)
La palabra `break` también funciona con el loop `for` 

````python
# en el siguiente ejemplo loop while es infinito porque el valor de la expresion es fijo: True 
# si la expresión después de "while" es (1>0) es también infinita
# Tenga cuidado con while infinita!
i=0
while (False):
    print(i)
    i=i+1
````

````python
# para terminar loop infinita
# se puede definir la condicion de terminar usando combinación: if,else y la palabra "break"
i=0
while (True):
    print(i)
    i=i+1
    if (i>5):
        break
````

## Funciones personales
Hasta ahora solo se han mostrado las funciones predefinidas; el usuario puede construir su propia función. Una función es un bloque de código que solo se ejecuta cuando se llama, una funcion puede recibir argumentos y devolver el resultado.

En Python, la construción de una función tiene la siguiente estructura:

    def nombre_de_funcion(argumento o argumentos separados con coma):  
        declaración
        return

Nótese que:
* nombre_de_funcion puede tener cualquier nombre (ciertas condiciones se explican más adelante).
* La indentación es obligatoria después de una declaración o un código que va a ser ejecutado (4 espacios).
* La función puede devolver un valor o más valores (variables) pero no es obligatorio.
* La función tiene que ser definida antes de ser llamada.

Entender esta estructura es más fácil con un ejemplo:

````python
# en este ejemplo se define una función con nombre "func_saludo"
# que recibe un argumento (variable) 
# e imprime un mensaje "hola, 'variable'"

def func_saludo(variable):
        print("hola,",variable,'!')
````

````python
# si la función es definida puede ser llamada (nunca al contrario)
# la variable puede una cadena de letras, por ejemplo 'Roger'
func_saludo('Roger')
````

````python
# la variable puede ser de cualquier tipo, también numérico
# que en este particular ejemplo no tiene mucho sentido
func_saludo(1000)
````

Una función en su bloque pude incluir la creación de nuevas variables, la ejecución de loops (`for` o `while`) o condiciones, y puede devolver uno o más valores.

````python
# en el siguiente ejemplo función que se llama func_suma_rango
# suma todos los números adentro de un rango definido usando: range(1,num_entero)

def func_suma_rango(num_entero):          # definición de función que recibe un número entero
    resultado_suma=0                      # creación de una variable resultado para guardar el valor
    for i in range(num_entero):           # ejecución de loop for
        resultado_suma=resultado_suma+i   # resultado_suma acumula suma de todos valores de rango
    return resultado_suma                 # función devuelve resultado_suma
````

````python
resultado_final=func_suma_rango(1000)     # a la función resultado_suma pasa número entero con valor 1000 
                                          # y devuelve valor que va a ser asignado a nueva variable: resultado_final
                                          # notése que resultado_suma y resultado_final pueden tener diferentes nombres
print(resultado_final)                    # imprimir resultado de esta funcion    
````

Muchas veces es necesario verificar si los parámetros que entran la función cumplen ciertos requisitos. Si el bloque de función requiere que el parámetro sea un número entero, el parámetro  no puede ser de tipo cadena o flotante, como en el ejemplo anterior.

````python
# la función anterior es modificada para verificar si el argumento es número entero
# porque range() acepta solo números enteros
def func_suma_rango(num_entero):              # definición de función que recibe un número entero
    if isinstance(num_entero,int):            # verificación de si argumento es entero `si True`
        resultado_suma=0                      # creación de una variable resultado para guardar el valor
        for i in range(num_entero):           # ejecución de loop for
            resultado_suma=resultado_suma+i   # resultado_suma acumula la suma de todos valores de rango
        return resultado_suma                 # función devuelve resultado_suma
    else:                                     # si isinstance(num_entero,int) es False ejecuta siguiente:
        print('argumento tiene que ser numero entero') # si argumento número no es entero sale información
````

````python
func_suma_rango(10) # si el argumento es número entero,la función ejecuta normalmente
````

````python
func_suma_rango(10.0) # si el argumento es número flotante, la función imprime mensaje de error
````

````python
func_suma_rango('Roger') # si el argumento es cadena de letras, la función imprime mensaje
````

````python
# en el siguiente ejemplo la función dividir_multiplicar acepta dos valores
# y devuelve dos resultados de multiplicación y división de estos dos números
# pero si el segundo parámetro es igual zero la función devuelve 

def dividir_multiplicar(a,b):
    resultado_multi=a*b
    if (b!=0):
        resultado_div=a/b
        return resultado_multi, resultado_div
    else:
        resultado_div=None
        print('No se puede dividir por zero!')
        return resultado_multi, resultado_div
````

````python
result_mult , result_div = dividir_multiplicar(20,0)
print(result_mult, result_div)
````

La palabra-clave `None` es usada para definir un valor nulo o ningun valor.

Python tiene una secuencia especial para manejar excepciones (errores) como la división por cero. 
Esta secuencia se llama: `try except`.
Las excepciones son errores que ocurren durante la ejecución del programa. Python no le dirá sobre errores como, por ejemplo, de sintaxis (fallas gramaticales), sino que se detendrá inmediatamente. 


El bloque de secuencia `try except` es el siguiente:

    try:   
        declaración, en caso que NO haya errores  
    except:  
        declaración, en caso que haya errores
        
Nótese que:
* `try`, `except` son palabras reservadas y no se pueden cambiar.
* La indentación es obligatoria después de una declaración o código que va a ser ejecutado.

Entender esta estructura es más facil con un ejemplo. Abajo se presenta la misma función `dividir_multiplicar()` con la implementación de `try/except`

````python
def divide_multiply_two_numbers_2(a,b):
    result_multiplication=a*b
    try:
        result_division=a/b
        return result_multiplication, result_division
    except:
        result_division=None
        print('No se puede dividir por zero!')
        return result_multiplication, result_division
````

````python
result_mult, result_div= divide_multiply_two_numbers_2(2,0)
print(result_mult, result_div)
````

#### Hay ciertas reglas para nombres de funciones de Python:
* Debe escribir el nombre de la función Python con todos los caracteres en minúsculas.
* No use caracteres en mayúsculas al nombrar una función en python.
* Use underscore `_` entre las palabras en lugar de espacio al nombrar una función.

## Formateo de los resultados
La funcion `print()` permite hacer el formateo de su argumento en la forma más avanzada si es necesario:
una de las formas más comunes es usando la letra `f` adentro de la función `print()`
que indica al intérprete que tiene que formatear el texto y otros argumentos. Después de `f` sigue la cadena de letras en apóstrofo único o doble, con las variables incluidas en corchetes, como `{variable}`.

La manera de formatear es la siguiente:
```
print(f'cadena de letras {variable1}, cadena de letras {variable2}...') 
```

Para imprimir información sobre número pi, se puede usar el formateo de texto:

````python
numero_pi=3.141592653589793
print(f'Numero pi que valor aproximado es {numero_pi} es la relación entre la longitud de una circunferencia y su diámetro en geometría euclidiana...')
````

Para imprimir información sobre número pi solamente con dos números decimales se puede usar
formateo con corchetes, incluyendo `:nf`, donde `n` es el número de dígitos decimales.

````python
numero_pi=3.141592653589793
print(f' El valor de numero pi con la precision de seis numeros flotantes es  {numero_pi:.6f} y con la precision de dos numeros flotantes es {numero_pi:.2f}')
````

La funcion `print()` también permite decidir si el nuevo resultado va ocurrir en una nueva línea o va seguir con la misma línea del resultado anterior

`print()` y argumento "end=" permite tres opciones
* `print(end='\n')`: después de imprimir un resultado, el inicio del resultado nuevo va a aparecer en la nueva línea ('\n' es predeterminado)
* `print(end=' ')`: después de imprimir un resultado, el inicio del resultado nuevo va a aparecer en el fin del resultado anterior
* `print(end='\r')`: después de imprimir un resultado, el inicio del resultado nuevo va a aparecer en la misma posición inicial del resultado anterior

````python
# ejemplo de usar print(end='\n')
for i in range(5):
    print(i) # por eso el nuevo resultado aparece en nueva línea
````

mismo como: `print(i,end='\n')`, porque `end='\n'` es predeterminado

````python
# ejemplo de usar print(end='\n')
for i in range(5):
    print(i,end='\n')
````

`end='\n'` dice: "después de imprimir ir a la nueva línea"

````python
# ejemplo de usar print(end=' ')
for i in range(5):
    print(i,end='') # cada resultado nuevo de función print() continua en la misma línea
````

`end=' '` dice: "después de imprimir mantenerse en la misma línea"

Tambien puede usar `print(end='\n')` sin valor, imprimiendo una cadena vacía yendo a la siguiente línea después

````python
print(1,end=' ')
print(2,end=' ')
print(end='\n')
print(1,end=' ')
print(2,end=' ')
print(3,end=' ')
````

Finalmente, último caso: `end='\r'`

````python
# repetido ejemplo de usar print(end='\r')
# usando biblioteca time se puede hacer pausa después de cada ejecución nueva de la loop 
# que permite ver el resultado
# el tiempo de la pausa se puede definir como argumento de time.sleep(1) 
import time
for i in range(10):
    print(i,end='\r')
    time.sleep(1)
````
