Para un dispositivo óptico particular que consiste en un medio no lineal dentro de una cavidad óptica, esto sólo ocurre cuando la intensidad I se encuentra dentro de un cierto rango que está dado por la ecuación: 

$$f(x)=(x+1)^2 - c(x-1) < 0 $$

donde $x$ es la intensidad del campo eléctrico y $c$ es la fuerza de la absorción del 
medio no lineal. 

## 1. Cree una función llamada `f_cost` que reciba dos parámetros (x,c) y solo devuelva la función f(x) como se mencionó anteriormente.

````python
import numpy as np
def f_cost(x,c):
   #x es la intensidad del campo eléctric y c es la fuerza de la absorción del medio No lineal
  return (x+1)**2 - c*(x-1)
````

## 2. Usando funcion `f_cost` haz un gráfico con seis curvas para c=5,10,15,20,25,30 y con x entre -10,30 (100 puntos)
* hacer que cada curva sea de un color diferente
* añadir etiquetas
* Para cual `c` esta funcion no tiene raices?

````python
import matplotlib.pyplot as plt
import numpy as np
import scipy.optimize as optimize
x = np.linspace(-10,30,100)
y0 = f_cost(x,5)  #probamos la funcion para cada valor de c
y1 = f_cost(x,10)
y2 = f_cost(x,15)
y3 = f_cost(x,20)
y4 = f_cost(x,25)
y5 = f_cost(x,30)
fig,ax = plt.subplots(1,1)
ax.plot(x,y0,label='funcion cuando c=5')
ax.plot(x,y1,label='funcion cuando c=10')
ax.plot(x,y2,label='funcion cuando c=15')
ax.plot(x,y3,label='funcion cuando c=20')
ax.plot(x,y4,label='funcion cuando c=25')
ax.plot(x,y5,label='funcion cuando c=30')
ax.legend()
ax.grid('True')
````

````python
#INTERPOLACION
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import scipy.optimize as optimize
from scipy.interpolate import interp1d
data = pd.read_csv('data_pressure_vs_temparature.csv')
print(data)
#lo convertimos en array
data_numpy = data.to_numpy()
#comprobamos que es un array
print(type(data_numpy))
#dividimos las columnas
temperature = data_numpy[:,0]
pressure = data_numpy[:,1]
ax.plot(temperature,pressure,label='grafica temperaturavspresion')
plt.show()
#interpolamos
interpolary = interp1d(temperature,pressure,kind = 'cubic')
xdata = np.linspace(temperature[0],temperature[30],100) #interpolamos 100 puntos
fig,ax = plt.subplots(1,1)
ax.scatter(temperature,pressure,label='data')

ax.plot(xdata,interpolary(xdata),'r',label='interpolate cubic')
plt.title('Pinterp vs  Tinterp')
ax.legend()
plt.show()
````

# Ejercicio III

El `polilogaritmo` que recibe dos parametros (https://www.mpmath.org/doc/current/functions/zeta.html) es una funcion complicada de calcular 

````python
import mpmath
import math
import numpy as np
import time
````

#### Para un valor se puede obtener rápidamente el resultado


````python
mpmath.polylog(3.0/2.0, 0.5)
````

#### Para calcular 50 puntos uno necesita mas tiempo

````python
timestamp1 = time.time()
xgrid1=np.linspace(0.0,1.0,50)

poly1=[]
for x in xgrid1:
    poly1.append(mpmath.polylog(3.0/2.0,x))

    
timestamp2 = time.time()
print (f"Esta operacion tomo : {(timestamp2 - timestamp1):.2f} segundos")
````

### Aqui es el plot (solo para mostrar el comportamiento)

````python
import matplotlib.pyplot as plt
import numpy as np
fig,ax=plt.subplots(1,1)
ax.plot(xgrid1,poly1,'.r',label='calculated points using Taylor exp.')
ax.set_xlabel('xgrid1')
ax.set_ylabel('poly1')

ax.legend()
ax.grid('True')
````
