## Matlab

Importar la biblioteca matplotlib.pyplot es tan simple como importar numpy o cualquier otra biblioteca de Python: 

`import matplotlib.pyplot as plt`

Donde "plt" es un alias que hace que el uso de la biblioteca sea mucho más conveniente.

Hay varias formas de usar matplotib y aquí se presentará una, muy flexible.

En el aprendizaje automático es común utilizar polinomios como parte de un método llamado Regresión que modela la relación entre variables. Usando numpy y matplotlib podemos visualizar fácilmente un polinomio (en este momento se hará énfasis solo en la visualización del polinomio, no en cómo encontrarlo).

Primero se puede crear una matriz numpy que represente la variable independiente $x$, y luego la variable dependiente $y$ (polinomio) que corresponde a una función de $f(x)$.

La función np.linspace() devuelve espacios numéricos de manera uniforme con respecto al numero de muestras.

np.linspace() con los tres parámetros más importantes tiene la siguiente sintaxis:

`np.linspace(start,stop,num)` donde

* start: inicio del intervalo
* stop : fin del intervalo
* num : número de muestras (número entero)

La función semejante np.arrange() devuelve espacios numéricos de manera uniforme con respecto al intervalo.

`np.arange(start,stop,interval)` donde

* start: inicio del intervalo
* stop : fin del intervalo
* interval : significa el espacio entre los intervalos

En el siguiente ejemplo se creará numpy array $x$ que represente la variable independiente, y usando función matemática la variable dependiente $y$ que corresponde a una función de $y=f(x)$ (por ejemplo, polinomio de tercer orden).

```python
import numpy as np  
x=np.linspace(-4.0,6.0,50)  # o por ejemplo usando np.arange: x=np.arange(-4.0,6.0,0.1)
y=4+2*x+3*x**2-0.7*x**3  
```

Luego de tener ambos vectores definidos, con matplotlib se puede crear una figura

`fig,ax = plt.subplots(1,1,figsize=(6,4))`

* `fig` (figure) es un contenedor y se pueden trazar datos sobre él.

* `ax` (axis) es el área real en la que se trazan los datos. Un `ax` representa una gráfica dentro de la figura.

* `plt.subplots(1,1,figsize=(6,4))` se crea un contenedor `fig` con tamaño de 6x4 pulgadas con solamente una figura gráfica (matriz de 1x1).

* `x=np.linspace(-4.0,6,50):` crear un vector con valores que estén igualmente espaciados entre -4:6. El número de elementos es 50. 
* `y=4+2*x+3*x**2-0.7*x**3: ` y es una función de x (aquí polinomio de tercer orden)

Para visualizar los dos vectores `x`, `y` se puede usar la sintaxis:

* `ax.plot(x,y)`    (línea)

Para hacer el plot más completo se puede añadir el título a la leyenda de la figura, y guardar la figura en el formato 'jpg'.

El código completo de este ejemplo particular se presenta a continuación:

```python
import numpy as np  
x=np.linspace(-4.0,6,50)  
y=4+2*x+3*x**2-0.7*x**3 

import matplotlib.pyplot as plt
fig, ax = plt.subplots(1,1,figsize=(6,4))            # creamos una grafica adentro de figura de tamano 6x4 pulgas
ax.plot(x,y,'.r',label='4+2x+3x^2-0.7x^3')           # curva, linea, punteada, color rojo '.r'
                                                     # nombre de la curva: '4+2x+3x^2-0.7x^3'

ax.set_title('polynomio')                            # titulo
ax.set_xlabel('x')                                   # nombre del x-axis
ax.set_ylabel('y')                                   # nombre del y-axis
ax.legend()                                          # mostrar nombre de la curva
ax.grid('True')                                      # hacer grid
fig.savefig('polynomio.jpg')                         # guardar plot como imagen tipo `jpg`
```

Es fácil trazar varios polinomios en la misma gráfica

```python
import numpy as np
x=np.linspace(-4.0,6,50)
y0=4+2*x+3*x**2-0.7*x**3
y1=4+3*x+2*x**2
y2=4+5*x

import matplotlib.pyplot as plt
fig, ax = plt.subplots(1,1,figsize=(6,4))

# primera curva tipo linea
ax.plot(x,y0,'.r',label='4+2x+3x^2-0.7x^3')           # linea punteada, color rojo ('.r')

# secunda curva tipo linea
ax.plot(x,y1,'-b',lw=3,label='4+2x+2x^2')            # linea solida, 
                                                     # color azul (b), 
                                                     # grosura: lw=3
                                                     # nombre: '4+2x+2x^2'
# tercera curva tipo linea
ax.plot(x,y2,'--g',lw=2,label='4+5x')                # linea discontinua ('--')
                                                     # color verde (g)
                                                     # grosura: lw=2
                                                     # nombre: '4+5x'
ax.legend()
ax.set_title('polynomios')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.grid('True')
fig.savefig('polynomios.jpg')
```

`plt.subplots()` permite crear una matriz de las gráficas:   
`fig, ax=plt.subplots(1,4,figsize=(14,3))`  
(1,4) significa una fila con cuatro columnas de las gráficas. Se puede cambiar la propiedad de cada gráfica a través de los ejes ax[0],ax[1],ax[2],ax[3] (cuatro contando desde cero)

```python
# plots in loop
fig, ax=plt.subplots(1,4,figsize=(14,3))

ax[0].plot(x,y,'.r',label='4+2x+3x^2-0.7x^3')
ax[0].set_title('y=4+2x+3x^2-0.7x^3')
ax[0].grid('True')

ax[1].plot(x,y1,'-b',lw=3,label='y=4+2x+2x^2')
ax[1].grid('True')
ax[1].set_title('y=4+2x+2x^2')

ax[2].plot(x,y2,'--g',lw=2,label='y=4+5x')
ax[2].grid('True')
ax[2].set_title('y=4+5x')

ax[3].plot(x,y0+y1,'-k')
ax[3].grid('True')
ax[3].set_title('y=y0+y1')
```

## Visualización de distribuciones

Para visualizar una distribución es mejor usar un `scatter` plot. La sintaxis es similar, excepto que la `scatter` tiene una opción de un tamaño de los puntos.

```python
import numpy as np
x=np.random.normal(0,0.2,500)
y=np.random.normal(0,0.2,500)
import matplotlib.pyplot as plt
fig, ax = plt.subplots(1,1,figsize=(6,4))

# primera curva tipo scatter
ax.scatter(x,y,c='r',s=30,label='xy') # curva punteada (scatter)
                                      # tamaño de los puntos, s=30
                                      # color de los puntos rojos, 'r' (red)
                                      # nombre de los puntos, 'xy'


ax.legend()
ax.set_title('data points')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.grid('True')
ax.set_xlim(-1,1)                  # límites de eje x
ax.set_ylim(-1,1)                  # límites de eje y

fig.savefig('data_points.jpg')
```

El tamaño de los puntos también puede ser una función:

```python
import numpy as np
N = 100
sigma=0.3
mu=0.0
x=np.random.normal(mu,sigma,N)
y=np.random.normal(mu,sigma,N)

import matplotlib.pyplot as plt
fig, ax = plt.subplots(1,1,figsize=(6,4))

colors = np.random.rand(N)
areas = (30 * np.random.rand(N))**2  
ax.scatter(x,y,c=colors,s=areas,label='xy')  # curva punteada con colores y áreas de los puntos aleatorios

ax.legend()
ax.set_title('data points colors, areas')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.grid('True')
ax.set_xlim(-1,1)                             # límites de eje x
ax.set_ylim(-1,1)                             # límites de eje y

fig.savefig('data_points_colors.jpg')
```

## Visualización de imágenes

Muchos algoritmos de aprendizaje automático (clasificación) requieren trabajar con imágenes. Antes de introducir los datos en un algoritmo es útil visualizar las imágenes para tener una mejor idea de su contenido. 

El módulo de imágenes en la biblioteca `matplotlib.pyplot` se utiliza para trabajar con imágenes en Python. El módulo de imagen también incluye dos métodos útiles que son:

* `imread` que se utiliza para leer imágenes 
* `imshow` que se utiliza para mostrar la imagen

En el siguiente ejemplo, el programa lee una imagen usando `matplotlib.image.imread()` y muestra esa imagen usando `matplotlib.image.imshow()`. Por ejemplo, las imágenes de limones son guardadas en una carpeta "data".

```python
! ls data/
```

```python
image=plt.imread('data/lemon1.jpg')    # función plt.imread(ruta) convierte imagen a una matriz
fig,ax=plt.subplots(1,1,figsize=(5,3)) # función plt.subplots() prepara una figura
ax.imshow(image)                       # función plt.imshow(img) muestra la imagen
```

Se puede ver fácilmente que una imagen importada no es más que una matriz de números (numpy array).

```python
type(image)
```

```python
# fragmento de esta gran 3D matriz
image[0:10,0:10,0]
```

Análogamente al ejemplo anterior, las imágenes también se pueden mostrar en forma de matriz con tamaño de una fila y cuatro columnas.

```python
fig, ax = plt.subplots(1, 4, figsize=(10, 2))
for i in range(4):                               # loop `for` incrementando `i` (0,1,2,3) 
    image=plt.imread('data/lemon'+str(i)+'.jpg') # la ruta para imagen cambia para cada imagen  
    ax[i].imshow(image,aspect="auto")            # cada gráfica visualiza la imagen distinto
    ax[i].set_axis_off()                         # los ticks no son necesarios para las imágenes 
```

Las mismas imagenes también se pueden mostrar en forma de matriz con dos filas y dos columnas:

```python
fig, ax = plt.subplots(2, 2, figsize=(7, 5))
k=0                                                 # declaración de contador `k`
for i in range(2):                                  # loop 1 `for` incrementando `i` (0,1)           
    for j in range(2):                              # loop 2 `for` incrementando `j` (0,1)
        k=k+1                                       # incremento de contador `k`
        image=plt.imread('data/lemon'+str(k)+'.jpg')# la ruta para imagen cambia para cada imagen
        ax[i,j].imshow(image,aspect="auto")         # ax es una matriz 2D: ax[i,j]
        ax[i,j].set_axis_off()                      # los ticks no son necesarios para imágenes 
```

Si se quiere, es posible elegir aleatoriamente las imágenes a mostrar usando la función np.random.randint()

```python
rand_idx=np.random.randint(0,10,4)
fig, ax = plt.subplots(1, 4, figsize=(12, 4))
k=0
for i in rand_idx:
    image=plt.imread('data/lemon'+str(i)+'.jpg')
    ax[k].imshow(image)
    ax[k].set_axis_off()
    k=k+1
```

## Conversión de imágenes a escala de grises

Una imagen en color es una matriz 3D. El aprendizaje automático a veces requiere la información sobre el objeto de color, pero a veces el color es una información redundante. Por ejemplo, un color de una fruta es una información importante para que el algoritmo (o humano) clasifique correctamente la fruta. Pero, por ejemplo, una imagen en color de un dígito no lleva más información que un color gris del mismo dígito.

```python
# a veces el color es importante (clasificación de las frutas)

import matplotlib.pyplot as plt
fig,ax=plt.subplots(1,1,figsize=(20,10))
ax.imshow(plt.imread('imgs/fruits_labels.png'))
ax.set_axis_off()
```

```python
# a veces el color de la imagen no es importante (fotos de números)

import matplotlib.pyplot as plt
fig,ax=plt.subplots(1,1,figsize=(10,5))
ax.imshow(plt.imread('imgs/house_number.jpg'))
ax.set_axis_off()
```

Una imagen importada por plt.imread() es una matriz numpy 3D que se llama RGB (***R***ed, ***G***reen, ***B***lue).

# $\color{red}{\text{Para el ilustrador:}}$

```python
image=plt.imread('imgs/rgb.png')
fig,ax=plt.subplots(1,1,figsize=(12,6))
ax.imshow(image)
ax.set_axis_off()
```

Lo que significa que hay tres capas de matrices 2D.  
Por ejemplo, se puede inspeccionar la imagen color de este dígito y ver que tiene tres dimensiones. 

```python
img_color=plt.imread('imgs/house_number.jpg')
fig,ax=plt.subplots(1,1)
ax.imshow(img_color)


print(image.shape)
```

Las dos primeras dimensiones (314, 850, 1000) son el ancho y el alto de una imagen, y la tercera dimensión (3) se refiere a las tres capas. Se puede mostrar un fragmento de cada capa.

*Como se mencionó anteriormente, el atributo `shape` de una matriz devuelve una tupla, que es otro tipo de datos en Python. La tupla está entre paréntesis `()` (a diferencia de la lista que está entre corchetes `[]`). La tupla no es mutable, está protegida de cambios. ¿Por qué? Porque la forma (`shape`) de una array es la propiedad de esta array y no tendría sentido cambiar ninguno de esos números, a menos que se cambie la forma de la matriz en sí.

```python
print('primeras 5 filas y 5 columnas de la primera capa')
print(img_color[0:5,0:5,0])
print()
print('primeras 5 filas y 5 columnas de la segunda capa')
print(img_color[0:5,0:5,1])
print()
print('primeras 5 filas y 5 columnas de la tercera capa')
print(img_color[0:5,0:5,2])
```

Con el fin de acelerar el aprendizaje automático cuando el color de la imagen no es importante se pueden reducir las tres capas de imagen en color a una sola capa que se llama imagen en escala de grises. Puede hacerse promediando las tres capas sobre la última dimensión usando `np.average()` (Recuerde que se puede acceder al último elemento de una matriz numpy utilizando el índice `-1`)

```python
img_gris=np.average(img_color,-1) # promedios valores sobre la última dimensión de la 3D numpy array
fig,ax=plt.subplots(1,1)
ax.imshow(img_gris)
```

La imagen RGB reducida a una capa se llama imagen en escala de grises y tiene solamente dos dimensiones.


```python
img_gris.shape
```

El mismo efecto de convertir una imagen de color a una imagen en escala de grises se logra usando un software de tercera parte llamado `Pillow`. A modo de demostración: 

```python
from PIL import Image
img_gris_pillow = Image.open('imgs/house_number.jpg').convert('L')
fig,ax=plt.subplots(1,1)
ax.imshow(img_gris_pillow)
```

*Convertir imágenes de color a la escala de grises es útil en el aprendizaje automático, ya que reduce la complejidad del problema. Puede llevarse a cabo para imágenes de dígitos o letras. Por ejemplo, para una letra 'A' el color no tiene significado semántico: una 'A' es una 'A' ya sea roja o verde. Y no es raro tener una 'A' roja. Sin embargo, el color a veces lleva información importante sobre un objeto, por ejemplo, un gato puede ser blanco, negro o marrón; nunca va a ver un gato verde o rojo.

## Aplanamiento de imágenes

Cuando se desea aplanar la imagen se debe convertir una matriz en un vector y, por ejemplo, guardarla como una fila de números, en una tabla grande numérica donde cada fila coresponda a una imagen. Se desarrollará este ejemplo mientras se trabaja con otra biblioteca de Python denominada Pandas.

* Convertir una matriz en un vector se llama aplanar (flatten) la matriz.
* En realidad la numpy array en la forma de matriz numérica aplanada es un dato de entrada a la red neuronal.

La función `np.reshape()` se utiliza para remodelar una matriz numpy sin cambiar los datos de la misma. Es una práctica muy común remodelar matrices con el fin de hacerlas compatibles para cálculos posteriores. [13]  

La sintaxis es la siguiente:
`numpy.reshape(matriz, (newshape))`

```python
img_vector = np.reshape(img_gris, (img_gris.size))
print(img_vector.shape)
```

De la misma manera, usando `np.reshape()` se remodela la matriz a sus dimensiones originales.

```python
img_matrix=np.reshape(img_vector, (1000, 1000))
```

```python
fig,ax=plt.subplots(1,1)
ax.imshow(img_matrix)
```
